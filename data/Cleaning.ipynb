```python
import pandas as pd
```

Data Link: https://www.kaggle.com/datasets/utsavdey1410/food-nutrition-dataset/data

- This is the food data group 1 set

```python
food = pd.read_csv('FOOD-DATA-GROUP1.csv')
```

```python
food.columns
```

```python
mcdonalds = food[food.food.str.contains('mcdonalds')]
```

```python
mcdonalds
```

```python
sandwich = ['cheeseburger mcdonalds', 'quarter pounder mcdonalds', 'big mac mcdonalds',
           'mcchicken mcdonalds', 'filet o fish mcdonalds',
           'egg mcmuffin mcdonalds', 'sausage mcmuffin mcdonalds']
```

```python
mcd_sandwich = mcdonalds[mcdonalds.food.isin(sandwich)].reset_index()
```

```python
mcd_sandwich = mcd_sandwich[['food', 'Caloric Value', 'Fat', 'Carbohydrates', 'Protein']]
```

```python
mcd_sandwich
```

```python
mcd_sandwich['food'] = mcd_sandwich.food.str.rstrip('mcdonalds')
```

```python
mcd_sandwich['food'] = mcd_sandwich.food.str.title()
```

```python
mcd_sandwich = mcd_sandwich.rename(columns={'Caloric Value':'Calories'})
```

```python
mcd_sandwich
```

```python
mcd_sandwich.to_csv('mcdonalds.csv', index=False)
```
